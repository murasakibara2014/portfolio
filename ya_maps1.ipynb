```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
service = Service(ChromeDriverManager().install())  # Используем Service для указания пути к драйверу
driver = webdriver.Chrome(service=service)

url = "https://yandex.ru/maps/org/stolovaya/121590504460/?indoorLevel=1&ll=39.890182%2C57.624362&z=16.67"

driver.get(url)
```

```python
close_button = driver.find_element(By.XPATH, '//div[@class="business-header-rating-view__text _clickable"]')
close_button.click()
```

```python
# business-reviews-card-view__review
```

```python
def scroll():

    from time import sleep

    elements = driver.find_elements(By.CLASS_NAME, 'business-reviews-card-view__review')
    count_el = len(elements)
    count_el2 = 0

    while count_el != count_el2:
        count_el = len(elements)
        print(count_el)
        last_element = elements[-1]
        driver.execute_script("arguments[0].scrollIntoView(true);", last_element)
        sleep(1)
        elements = driver.find_elements(By.CLASS_NAME, 'business-reviews-card-view__review')
        count_el2 = len(elements)
        print(count_el2)
        
    return count_el2

count_el = scroll()
```

```python
# копируем html код страницы

data = driver.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data)
```

```python
info_block = soup.find_all(class_='business-reviews-card-view__review')
len(info_block)
```

```python
a = info_block[1]
a
```

```python
comment = a.find(class_='business-review-view__body-text').text
comment
```

```python
name = a.find(class_='business-review-view__link').text
name
```

```python
level = a.find(class_='business-review-view__author-caption').text
level
```

```python
ratings = a.find('span', itemprop='reviewRating', itemtype='http://schema.org/Rating')
rating = ratings.find_all('meta')
rating = rating[2]['content']
rating
```

```python
date = a.find(class_='business-review-view__date').text
date
```

```python
data_dict = {'name':[], 'level':[], 'rating':[], 'date':[], 'comment':[]}

for i in range (0, count_el):
    
    try:
        name = info_block[i].find(class_='business-review-view__link').text
        data_dict['name'].append(name)
    except:
        data_dict['name'].append('Нет данных')
    
    try:
        level = info_block[i].find(class_='business-review-view__author-caption').text
        data_dict['level'].append(level)
    except:
        data_dict['level'].append('Нет данных')
    
    try:
        ratings = info_block[i].find('span', itemprop='reviewRating', itemtype='http://schema.org/Rating')
        rating = ratings.find_all('meta')
        rating = rating[2]['content']
        data_dict['rating'].append(rating)
    except:
        data_dict['rating'].append('Нет данных')
    
    try:
        date = info_block[i].find(class_='business-review-view__date').text
        data_dict['date'].append(date)
    except:
        data_dict['data'].append('Нет данных')
        
    try:
        comment = info_block[i].find(class_='business-review-view__body-text').text
        comment1 = comment.replace('\n', '')
        data_dict['comment'].append(comment1)
    except:
        data_dict['comment'].append('Нет данных')
```
